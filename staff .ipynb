# IQR for absent(only numeric column)


```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include='number')

# Calculating Q1, Q3, and IQR
iqr_df = numeric_cols.apply(lambda x: pd.Series({
    'Q1': x.quantile(0.25),
    'Q3': x.quantile(0.75),
    'IQR': x.quantile(0.75) - x.quantile(0.25)
}))

print(iqr_df.T)

```

# Perform Univariate, Bivariate, and Multivariate analysis to extract meaningful insights and identify relationships among variables.

```python
import pandas as pd

df = pd.read_csv("staff.csv")
df
```

```python
# Convert absent values to 0 and 1:
df['absent'] = df['absent'].replace({'False': 0, 'True': 1})
```

```python
# Save the cleaned DataFrame to a CSV file:
df.to_csv('cleaned_data.csv', index=False)
```

```python
# Reload the CSV with correct types (optional):

df = pd.read_csv('cleaned_data.csv', dtype={'absent': int})
df
```

```python
# Verify:

print(df['absent'].unique())
print(df.dtypes)
```

# 🔹 1. Univariate Analysis (one variable at a time)


```python
# absent column(only numeric column)

import seaborn as sns
import matplotlib.pyplot as plt

# Histogram of absent
sns.histplot(df['absent'], bins=3, kde=True)
plt.title("Distribution of Absence")
plt.xlabel("Absent (0 = Present, 1 = Absent)")
plt.tight_layout()
plt.show()
```

# ➤ department, shift (Categorical):


```python
# Countplot of department
sns.countplot(x='department', data=df)
plt.title("Department Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Countplot of shift
sns.countplot(x='shift', data=df)
plt.title("Shift Distribution")
plt.tight_layout()
plt.show()
```

# 🔹 2. Bivariate Analysis (relationship between two variables)


# ➤ absent by shift:



```python
sns.boxplot(x='shift', y='absent', data=df)
plt.title("Absence by Shift")
plt.tight_layout()
plt.show()
```

# ➤ absent by department:


```python
sns.boxplot(x='department', y='absent', data=df)
plt.title("Absence by Department")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 🔹 3. Multivariate Analysis (3+ variables)
Generally used heatmaps or scaterplots


# ➤ Heatmap of Absences grouped by department and shift:



```python
grouped = df.groupby(['department', 'shift'])['absent'].mean().unstack()
sns.heatmap(grouped, annot=True, cmap="YlGnBu")
plt.title("Average Absence by Department & Shift")
plt.tight_layout()
plt.show()
```
